```python
import pandas as pd
from google.colab import data_table
```

```python
data = pd.read_csv('PS_20174392719_1491204439457_log.csv')
data.head(1000)
```

```python
data.columns[data.isna().any()].tolist()
```

```python
data.drop(['isFlaggedFraud','nameOrig','nameDest', 'amount'], axis = 1, inplace = True)
data
```

```python
data = pd.get_dummies(data, columns = ['type'])
data
```

```python
data.drop(['type_CASH_IN', 'type_CASH_OUT', 'type_DEBIT', 'type_TRANSFER'], axis = 1, inplace = True)
data
```

```python
rat = len(data.loc[data['isFraud']==0])//len(data.loc[data['isFraud']==1])

df_1 = data.loc[data['isFraud']==1]
df_1 = df_1.loc[df_1.index.repeat(rat)]
train_n = pd.concat([data.loc[data['isFraud']==0], df_1]).sample(frac=1)

data = train_n.sample(n = 20000, random_state=1)
```

```python
X = data.drop('isFraud', axis = 1)
y = data['isFraud']
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X , y, test_size=0.3, random_state=42)


knn = KNeighborsClassifier(n_jobs = -1)

```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(knn, X, y, cv=5)

scores
```

```python
import numpy as np
n_neighbors_options = range(1, 21)
weights_options = ['uniform', 'distance']
metric_options = ['euclidean', 'manhattan']
algorithm_options = ['auto', 'ball_tree', 'kd_tree', 'brute']

results = []

for n_neighbors in n_neighbors_options:
    for weights in weights_options:
        for metric in metric_options:
          for algorithm in algorithm_options:
              knn = KNeighborsClassifier(n_neighbors=n_neighbors, weights=weights, metric=metric, algorithm=algorithm, n_jobs = -1)
              scores = cross_val_score(knn, X_train, y_train, cv=5, n_jobs=-1)
              results.append({
                  'n_neighbors': n_neighbors,
                  'weights': weights,
                  'metric': metric,
                  'algorithm': algorithm,
                  'average_accuracy': np.mean(scores)
              })

best_result = max(results, key=lambda x: x['average_accuracy'])


print(best_result)
```

```python
knn_best = KNeighborsClassifier(n_neighbors=4, weights = 'distance', metric = 'manhattan', algorithm = 'auto', n_jobs=-1)
knn_best.fit(X_train, y_train)

pred = knn_best.predict(X_test)

from sklearn.metrics import classification_report

print(classification_report(y_test, pred))

print(bias_variance_decomp(knn_best, np.array(X_train), np.array(y_train), np.array(X_test), np.array(y_test), loss='mse', num_rounds=10))
```
